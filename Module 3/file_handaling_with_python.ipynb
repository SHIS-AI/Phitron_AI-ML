```python
## class problem solov
## akta file a koita line ase, koita word ase, koita char ase, ta amra python deya problem solov korbo
from functools import reduce
with open("./test file/data.txt",'r') as file:
    strngs = file.readlines() ## ata deya amra file a koto line ase ta bole dite se
    print(f"line size: {len(strngs)}") 

    word = list(map(lambda x: len(x.split()),strngs)) ## ata deya amra file a koto word ase ta count korbo.
    count_word = reduce(lambda x,y: x+y,word)
    print(f"file word size: {count_word}")

    char = list(map(lambda x : len(x.strip().replace(" ","")),strngs)) ## ata deya amra file koto char ase ta count korlam
    count_char = reduce(lambda x,y:x+y,char)
    print(f"file char size: {count_char}")

    with open("./test file/counter_of_string.txt","w") as f:
        f.write(f"line size: {len(strngs)}\n")
        f.write(f"file word size: {count_word}\n")
        f.write(f"file char size: {count_char}")



```
