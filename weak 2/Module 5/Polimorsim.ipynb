```python
class Camera:
    def __init__(self,name):
        self.name = name 
    def cupture(self):
        print("Capture a photo.")
```

```python
class Drone(Camera):
    
    def __init__(self,name,resulation): 
        super().__init__(name)
        self.resulation = resulation
    def cupture(self):
        print("Capture a drone photo.")
class DSLR(Camera): 
    def __init__(self,name,resulation):
        super().__init__(name)
        self.resulation = resulation
    def cupture(self): 
        print("Cupture a DSLR photo.")
class Mobile(Camera): 
    def __init__(self,name,resulation):
        super().__init__(name)
        self.resulation = resulation
    def cupture(self): 
        print("Cupture a mobile photo.")
        
```

```python
dslr = DSLR('dslr',200)
drone = Drone('drone',200)
phone = Mobile('mobile',200)
```

```python
dslr.cupture()
drone.cupture()
phone.cupture()
```
