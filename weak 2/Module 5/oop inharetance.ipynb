```python
## creating opbject
class Phone:
    battery_percentage = 0
    ## catagori
    catagori = 'Supper Electory'
    def __init__(self,model, battery, camera):
        self.model = model
        self.battery = battery
        self.camera = camera
    def cherging(self,charge):
        self.battery_percentage +=charge
        print(f"carent battery percentage {self.battery_percentage}")

    # photo cupturing
    def photo_cupturing(self,photo):
        if self.battery_percentage > 0 and photo < self.battery_percentage:
            self.battery_percentage -= photo
            print(f"You are {photo} photo click")
            print(f"carent battery percentage {self.battery_percentage}")
        else:
            print("power low")

    
```

```python
nokiya = Phone('A21 pro max', 7000, '200mx')
```

```python
class Smart_Phone(Phone):
    def __init__(self,model,barrery,camera,procesor):
        super().__init__(model,barrery,camera)
        self.procesor = procesor
    def fast_charging(self,hour): 
        print("Fast charging is process")
        super().cherging(hour)
```

```python
apple = Smart_Phone('iphone 17 pro max', 5000, '50mx','SD')
```
