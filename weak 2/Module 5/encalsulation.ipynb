```python
class Mobile: 
    def __init__(self,name,model,emi): 
        self.__name = name 
        self.model = model
        self.__emi = emi
    def emi_getter(self): 
        return self.__emi
    def name_getter(self): 
        return self.__name
    def name_setter(self,name):
        self.__name = name
    
    
        
```

```python
iphone = Mobile('Iphone','17 pro','121763518')
```

```python
iphone.emi_getter()
```
